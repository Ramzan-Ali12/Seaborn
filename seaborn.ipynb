```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event",markers=True)
plt.show()
```

Seaborn Histogram/DistPlot

```python
import pandas as pd
import seaborn as sns
df=pd.read_csv("E:\PYTHON Codes/diamonds.csv")
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df.head()
```

Horizontially plot

```python
sns.distplot(df['price'],color="green")
plt.show()
```

Vertically plot

```python
sns.distplot(df['price'],color="green",vertical=True)
plt.show()
```

```python
sns.distplot(df['price'],hist=False)
plt.show()
```

Horizontally Plot

```python
sns.distplot(df['price'],kde=False,color="purple",bins=5)
plt.show()
```

Vertically Plot

```python
sns.distplot(df['price'],kde=False,color="purple",bins=5,vertical=True)
plt.show()
```
